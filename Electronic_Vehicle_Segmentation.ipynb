```python
data3['PowerTrain'].replace(to_replace=['RWD','FWD','AWD'],value=[0, 1, 2],inplace=True)
data3['RapidCharge'].replace(to_replace=['No','Yes'],value=[0, 1],inplace=True)
```

```python
X = data3[['AccelSec','TopSpeed_KmH','Efficiency_WhKm','FastCharge_KmH', 'Range_Km', 'RapidCharge', 'Seats', 'PriceEuro','PowerTrain']]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
pca = PCA(n_components=9)
X_pca = pca.fit_transform(X_scaled)
df_pca = pd.DataFrame(X_pca, columns=['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6', 'PC7', 'PC8', 'PC9'])
df_pca.head()
```

```python
kmean = KMeans(n_clusters=4, init='k-means++', random_state=90)
kmean.fit(X_pca)
```

```python
KMeans(n_clusters=4, random_state=90)
```

```python
data3['clusters'] = kmean.labels_
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_pca, x='PC1', y='PC9', s=70, hue=kmean.labels_, palette='coolwarm', zorder=2, alpha=0.8)  
plt.scatter(x=kmean.cluster_centers_[:,0], y=kmean.cluster_centers_[:,1], marker="*", c="black", s=150, label="Centroids", edgecolor='white', zorder=3)

plt.xlabel('PC1', family='serif', size=13, weight='bold')  
plt.ylabel('PC9', family='serif', size=13, weight='bold')  
plt.xticks(family='serif')
plt.yticks(family='serif')
plt.grid(True, which='both', linestyle='--', linewidth=0.7, alpha=0.5, zorder=1)  

plt.legend(title='Clusters', fancybox=True, shadow=True, fontsize=10, title_fontsize=11, loc='upper right')
plt.title('K-Means Clustering Results', family='serif', size=16, weight='bold')

plt.show()
```
